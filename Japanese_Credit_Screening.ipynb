**PCA**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_sc=sc.fit_transform(data)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(data_sc, target, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
train_pred=lr.predict(x_train)
test_pred=lr.predict(x_test)
```

```python
print("Training accuracy: ", accuracy_score(train_pred, y_train))
print("Training accuracy: ", accuracy_score(test_pred, y_test))
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=15)
```

```python
data_pca=pca.fit_transform(data_sc)
```

```python
data_pca.shape
```

```python
x_train1, x_test1, y_train1, y_test1=train_test_split(data_pca, target, test_size=0.3)
```

```python
lr1=LogisticRegression()
lr1.fit(x_train1, y_train)
```

```python
train_pred1=lr1.predict(x_train1)
test_pred1=lr1.predict(x_test1)
```

```python
print("Training Accuracy: ", accuracy_score(train_pred1, y_train))
print("Training Accuracy: ", accuracy_score(test_pred1, y_test))
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train1, y_train1)
```

```python
train_pred2=svc.predict(x_train1)
test_pred2=svc.predict(x_test1)
```

```python
print("Training Accuracy: ", accuracy_score(train_pred2, y_train))
print("Training Accuracy: ", accuracy_score(test_pred2, y_test))
```

**Naive Bayes**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(data, target, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.fit_transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train_sc, y_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(nb.predict(x_train_sc), y_train))
print(accuracy_score(nb.predict(x_test_sc), y_test))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train_sc, y_train)
```

```python
print(accuracy_score(dt.predict(x_train_sc), y_train))
print(accuracy_score(dt.predict(x_test_sc),y_test))
```

```python
dt1=DecisionTreeClassifier(criterion='entropy')
dt1.fit(x_train_sc, y_train)
```

```python
print(accuracy_score(dt1.predict(x_train_sc), y_train))
print(accuracy_score(dt1.predict(x_test_sc),y_test))
```
